```python
def compute_cost(x,y,w,b):
    f = np.dot(w,x)+b
    cost = (f-y)**2
    return 1/(2*len(x))*sum(cost)
```

```python
compute_cost(x_train, y_train, 2 , 1)
```

```python
def compute_gradient(x,y,w,b):
    f = np.dot(w,x) + b
    dj_dw = sum((f-y)*x)/len(x)
    dj_db = sum((f-y))/len(x)
    return dj_dw, dj_db
```

```python


plt.plot(compute_gradient(x_train, y_train,2,1))
```
